```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

%matplotlib inline
```

```python
#import covid.csv file
covid = pd.read_csv('../data/covid.csv')
covid.tail()
```

```python
#drop rows with nulll values
covid = covid.dropna()
covid.head()
```

```python
#drop column
covid = covid.drop(columns=['Deaths per 100000'])

#rename column
covid=covid.rename(columns = {'Deaths/1000':'COVID-19'})
covid.tail()
```

```python
#import spanish flu.csv file
spanish_flu = pd.read_csv('../data/spanish_flu.csv')
spanish_flu.tail()
```

```python
#drop rows with nulll values
spanish_flu = spanish_flu.dropna()
spanish_flu.head(10)
```

```python
#only keep needed columns
spanish_flu = spanish_flu[['City', 'Deaths/1000']].copy()

#rename columns
spanish_flu=spanish_flu.rename(columns = {'Deaths/1000':'Spanish Flu'}).round(2)
spanish_flu.head(10)
```

```python
#merge df's
flu_deaths = pd.merge(covid, spanish_flu, on = 'City', how = 'inner')
flu_deaths.head()
```

```python
#data = df.reset_index().melt('date', var_name='a', value_name='b')
data = flu_deaths.melt('City', var_name='a', value_name='b')
print (data)
```

```python
#ax = sns.barplot(x="a", y="b", hue="City", data=data)
#order = ['Baltimore, MD','Boston, MA','Chicago, IL','Cleveland, OH','Los Angeles, CA','Nashville, TN','New York, NY',
#          'Philadelphia, PA','Pittsburgh, PA','St. Louis, MO','San Francisco, CA','Total']

flu_deaths.set_index('City').plot.barh()
plt.title('Spanish Flu Deaths versus COVID-19 Deaths')
plt.xlabel('Deaths/1000')
plt.ylabel('City')
plt.gca().invert_yaxis();
#plt.yticks(range(len(order)), order);
```
